```python
less_trip_df = trip_df[trip_df['ended_at'] < '2022-02-15']
```

```python
#set the geo=spatial data
data = [go.Scattermapbox(
            lat = less_trip_df['start_lat'] ,
            lon = less_trip_df['start_lng'],
            customdata = less_trip_df['rideable_type'],
            mode = 'markers',
            marker = dict(
                size = 7,
                color = 'red',
                opacity = .8,
            ),
          )]
#set the layout to plot
layout = go.Layout(autosize = False,
                   mapbox = dict(accesstoken=mapbox_access_token,
                                bearing=5,
                                pitch=1,
                                zoom=13,
                                center= dict(lat=40.721319,
                                             lon=-73.987130)
                                ),
                    width=900,
                    height=600, 
                    title = "Citi bikes in New York")
```

```python
import plotly.io as pio
fig = dict(data=data, layout=layout)
pio.show(fig)
```
